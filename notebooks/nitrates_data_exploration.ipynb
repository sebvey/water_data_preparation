```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.dates as md
```

```python
# Loads from local csv file
nitrates = pd.read_csv('../data/nitrates/saone_2011_2021.csv')
nitrates.day = pd.to_datetime(nitrates.day)
nitrates.head()
```

```python
# Gets the History for each station in a dcitionary
station_nitrates = {}

for station_id in nitrates.station_id.unique() :
    station_nitrates[station_id] = nitrates[nitrates['station_id'] == station_id][['day','measure']]
    station_nitrates[station_id].set_index('day',inplace=True)
```

```python
# Plots the History for each station

for station_id,df in station_nitrates.items() :
    
    with plt.style.context('fivethirtyeight'):

        plt.figure(figsize=(10,6))
        plt.plot(df.index,df.measure)

        plt.title(f'Station {station_id} : Nitrates History')
        plt.ylabel('Nitrates (mg(NO3)/L)')

        ax = plt.gca()
        ax.set_ylim(ymin=0,ymax=25)

        plt.show()
```
